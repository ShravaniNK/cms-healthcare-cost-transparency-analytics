```python
# Cell 1 — Imports
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 50)
```

```python
# Cell 2 — Load Beneficiary Summary files (one per year)
bene_2008 = pd.read_csv('../data/raw/2008_beneficiary_summary_sample1.csv')
bene_2009 = pd.read_csv('../data/raw/2009_beneficiary_summary_sample1.csv')
bene_2010 = pd.read_csv('../data/raw/2010_beneficiary_summary_sample1.csv')

print(bene_2008.shape, bene_2009.shape, bene_2010.shape)
bene_2008.head()
```

```python
# Cell 3 — Load claims files
inpatient = pd.read_csv('../data/raw/inpatient_claims_sample1.csv')
outpatient = pd.read_csv('../data/raw/outpatient_claims_sample1.csv')

print("Inpatient:", inpatient.shape)
print("Outpatient:", outpatient.shape)
inpatient.head()
```

```python
# Cell 4 — Check key columns exist as expected
# Beneficiary Summary key columns:
bene_cols = ['DESYNPUF_ID', 'BENE_BIRTH_DT', 'BENE_DEATH_DT', 'BENE_SEX_IDENT_CD',
             'BENE_RACE_CD', 'SP_STATE_CODE', 'BENE_COUNTY_CD',
             'SP_CHF', 'SP_CHRNKIDN', 'SP_CNCR', 'SP_COPD', 'SP_DEPRESSN',
             'SP_DIABETES', 'SP_ISCHMCHT', 'SP_OSTEOPRS', 'SP_RA_OA', 'SP_STRKETIA',
             'MEDREIMB_IP', 'BENRES_IP', 'MEDREIMB_OP', 'BENRES_OP']

print([c for c in bene_cols if c not in bene_2008.columns])  # should be empty

```

```python
# Claims key columns:
claims_cols = ['DESYNPUF_ID', 'CLM_ID', 'CLM_FROM_DT', 'CLM_THRU_DT',
               'PRVDR_NUM', 'CLM_PMT_AMT', 'ICD9_DGNS_CD_1']

print([c for c in claims_cols if c not in inpatient.columns])
print([c for c in claims_cols if c not in outpatient.columns])
```

```python
# Cell 5 — Add a YEAR column to each beneficiary file, then stack them
bene_2008['BENE_YEAR'] = 2008
bene_2009['BENE_YEAR'] = 2009
bene_2010['BENE_YEAR'] = 2010

beneficiary_all = pd.concat([bene_2008, bene_2009, bene_2010], ignore_index=True)
print(beneficiary_all.shape)
```

```python
# Cell 6 — Parse date columns
beneficiary_all['BENE_BIRTH_DT'] = pd.to_datetime(beneficiary_all['BENE_BIRTH_DT'], format='%Y%m%d')
beneficiary_all['BENE_DEATH_DT'] = pd.to_datetime(beneficiary_all['BENE_DEATH_DT'], format='%Y%m%d', errors='coerce')

inpatient['CLM_FROM_DT'] = pd.to_datetime(inpatient['CLM_FROM_DT'], format='%Y%m%d')
inpatient['CLM_THRU_DT'] = pd.to_datetime(inpatient['CLM_THRU_DT'], format='%Y%m%d')

outpatient['CLM_FROM_DT'] = pd.to_datetime(outpatient['CLM_FROM_DT'], format='%Y%m%d')
outpatient['CLM_THRU_DT'] = pd.to_datetime(outpatient['CLM_THRU_DT'], format='%Y%m%d')
```

```python
# Cell 7 — Add a claim year and claim type flag for later stacking
inpatient['CLAIM_YEAR'] = inpatient['CLM_FROM_DT'].dt.year
inpatient['CLAIM_TYPE'] = 'Inpatient'

outpatient['CLAIM_YEAR'] = outpatient['CLM_FROM_DT'].dt.year
outpatient['CLAIM_TYPE'] = 'Outpatient'
```

```python
# Cell 8 — Quick null check before deciding cleaning rules
print(beneficiary_all.isnull().sum()[beneficiary_all.isnull().sum() > 0])
print(inpatient['CLM_PMT_AMT'].isnull().sum())
print(outpatient['CLM_PMT_AMT'].isnull().sum())
```

```python
# Cell 9 — Save cleaned intermediate files for the next notebook (as CSV files)
beneficiary_all.to_csv('../data/processed/beneficiary_all.csv', index=False)
inpatient.to_csv('../data/processed/inpatient_clean.csv', index=False)
outpatient.to_csv('../data/processed/outpatient_clean.csv', index=False)
```

```python
# Cell 10 — Combine inpatient + outpatient claims into one claims table
# Keep only shared/common columns needed for cost analysis, plus claim-type flag
claims_cols = ['DESYNPUF_ID', 'CLM_ID', 'CLM_FROM_DT', 'CLM_THRU_DT',
               'PRVDR_NUM', 'CLM_PMT_AMT', 'CLAIM_YEAR', 'CLAIM_TYPE']

all_claims = pd.concat([
    inpatient[claims_cols],
    outpatient[claims_cols]
], ignore_index=True)

print(all_claims.shape)
all_claims.head()
```

```python
all_claims.shape
```

```python
# Cell 11 — Join claims to beneficiary demographic + chronic condition data
# Use BENE_YEAR == CLAIM_YEAR so each claim links to that year's beneficiary record
claims_with_bene = all_claims.merge(
    beneficiary_all,
    left_on=['DESYNPUF_ID', 'CLAIM_YEAR'],
    right_on=['DESYNPUF_ID', 'BENE_YEAR'],
    how='left'
)

print(claims_with_bene.shape)
claims_with_bene.head()
```

```python
# Cell 12 — Check for unmatched claims (beneficiary info missing after join)
unmatched = claims_with_bene['BENE_YEAR'].isnull().sum()
print(f"Claims with no matching beneficiary record: {unmatched}")
```

```python
# Cell 13 — Save the joined, analysis-ready dataset
claims_with_bene.to_csv('../data/processed/claims_with_beneficiary.csv', index=False)
```
